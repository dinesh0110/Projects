```python
plt.figure(figsize=(20,5))
ax=sns.barplot(x=list(df.groupby('category').sum().sort_values(by='total',ascending=False).index),
               y=list(df.groupby('category').sum().sort_values(by='total',ascending=False).total))
ax.set_xticklabels(ax.get_xticklabels(),rotation=40)
plt.xlabel('Category',fontsize=20, color='red')
plt.ylabel('Total', fontsize=20, color='red')
plt.title('Revenue of each category', fontsize=25, color='red')
plt.show()
print(df.groupby('category').sum()['total'].sort_values(ascending=False).head(5))
```

```python


#Highest quantity
plt.figure(figsize=(20,5))
ax=sns.barplot(x=list(df.groupby('category').sum().sort_values(by='quantity',ascending=False).index),
               y=list(df.groupby('category').sum().sort_values(by='quantity',ascending=False).quantity))
ax.set_xticklabels(ax.get_xticklabels(),rotation=40)
plt.xlabel('Category',fontsize=20, color='red')
plt.ylabel('Quantity', fontsize=20, color='red')
plt.title('Total quantity of each category', fontsize=25, color='red')
plt.show()
print(df.groupby('category').sum()['quantity'].sort_values(ascending=False).head(10))

```

```python
df['timestamp']= pd.to_datetime(df['timestamp'])
df['hours'] = df['timestamp'].dt.hour + df['timestamp'].dt.minute/60
df['date']= df['timestamp'].dt.date
df['date']=pd.to_datetime(df['date'])
df['day']= df['date'].dt.day_name()
```

```python
con_df=pd.concat([df, pd.get_dummies(df['day'])],axis=1)
con_df
```

```python
for i in range(len(con_df['quantity'])):
    con_df['Wednesday'][i]=con_df['quantity'][i] * con_df.loc[:,"Wednesday"][i]
    con_df['Thursday'][i]=con_df['quantity'][i] * con_df.loc[:,"Thursday"][i]
    con_df['Friday'][i]=con_df['quantity'][i] * con_df.loc[:,"Friday"][i]
    con_df['Saturday'][i]=con_df['quantity'][i] * con_df.loc[:,"Saturday"][i]
    con_df['Sunday'][i]=con_df['quantity'][i] * con_df.loc[:,"Sunday"][i]
    con_df['Monday'][i]=con_df['quantity'][i] * con_df.loc[:,"Monday"][i]
    con_df['Tuesday'][i]=con_df['quantity'][i] * con_df.loc[:,"Tuesday"][i]
```

```python
Daily_df=pd.DataFrame(con_df.groupby('category').sum()[['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']])
Daily_df
```

```python
# @title Sunday

from matplotlib import pyplot as plt
Daily_df['Sunday'].plot(kind='line', figsize=(8, 4), title='Sunday')
plt.gca().spines[['top', 'right']].set_visible(False)
```
